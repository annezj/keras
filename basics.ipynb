```python
# load libraries and set random seed
import tensorflow as tf
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.models import Sequential
import numpy
numpy.random.seed(7)
```

```python
# load pima indians dataset
dataset = numpy.loadtxt("/Users/annejones/Documents/nn/pima-indians-diabetes.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# quick examples of numpy array creation and indexing
# 1d list of data
data = [11, 22, 33, 44, 55]
# array of data
data = numpy.array(data)
print(data)
print(type(data))
print(numpy.shape(data))
print(data[1])
print(data[-1])
# 2d list of data
data = [[11, 22],
       [33, 44],
        [55, 66]]
data = numpy.array(data)
print(data)
print(type(data))
print(numpy.shape(data))
print(data[0,1])
print(data[0,])
```

```python
# numpy array slicing
# define array
data = numpy.array([11, 22, 33, 44, 55])
#print(data[1:2])
#print(data[-1:])
# 2d
data = numpy.array([[11, 22, 33],
[44, 55, 66],
[77, 88, 99]])
# separate data into all but last column and last column
X, y = data[:, :-1], data[:, -1]
print(X)
#print(y)
```

```python
# how to reshape arrays
print(data.shape[0])
print(data.shape[1])
data = numpy.array([11, 22, 33, 44, 55])
print(data)
# reshape to 1 column and 5 rows
data = data.reshape(data.shape[0], 1)
print(data)
```

```python
# reshape 2d array to 3d array
data = [[11, 22],
        [33, 44],
        [55, 66]]
# array of data
data = numpy.array(data)
print(data.shape)
# reshape
data = data.reshape((data.shape[0], data.shape[1], 1))
print(data.shape)
```

```python
# Now back to Keras
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', 
              optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
X = dataset[:,0:8]
Y = dataset[:,8]
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# calculate predictions
predictions = model.predict(X)
# round predictions
rounded = [round(x[0]) for x in predictions]
print(rounded)
```
